## Exercise 5 - Image processing

Add another of images to find out what happens when you only null one channel. 

```python
img = plt.imread('../images/IMG_0756.jpg')

# Make a figure with three axes in a single row
fig, axs = plt.subplots(1, 3, figsize=(12, 6))

# Copy of image for red, green and blue pixel components
red = img.copy()
green = img.copy()
blue = img.copy()


# Get rid of other components
red[:,:,[1,2]] = 0  # Null out green and blue
green[:,:,[0,2]] = 0  # Null out red and blue
blue[:,:,[0,1]] = 0  # Null out red and green


# Display each new image
axs[0].imshow(red)
axs[1].imshow(green)
axs[2].imshow(blue);
```

What happens to the image when you only null out one of the RGB channels? Make three more copies of the array and add another row of subplots to find out. 

```python
img = plt.imread('../images/IMG_0756.jpg')

# Make a figure with three axes in a single row
fig, axs = plt.subplots(2, 3, figsize=(12, 8))

# Copy of image for red, green and blue pixel components
red = img.copy()
green = img.copy()
blue = img.copy()
magenta = img.copy()
yellow = img.copy()
cyan = img.copy()

# Get rid of other components
red[:,:,[1,2]] = 0  # Null out green and blue
green[:,:,[0,2]] = 0  # Null out red and blue
blue[:,:,[0,1]] = 0  # Null out red and green
magenta[:,:,0] = 0  # Null out green and blue
yellow[:,:,1] = 0  # Null out red and blue
cyan[:,:,2] = 0  # Null out red and green

# Display each new image
axs[0,0].imshow(red)
axs[0,1].imshow(green)
axs[0,2].imshow(blue)
axs[1,0].imshow(magenta)
axs[1,1].imshow(yellow)
axs[1,2].imshow(cyan);
```
