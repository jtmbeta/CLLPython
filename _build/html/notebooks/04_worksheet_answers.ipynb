> Thirty days has September,\

## Exercise 4

Here's another good use of a pie chart.

<img src="../images/pacman.jpg" width=600/>

Write the code that makes this plot.

```python
import matplotlib.pyplot as plt

# Labels for each slice of the pie.
labels = [
    "Looks like pacman",
    "Doesn't look like pacman"
]

# Values representing slice sizes.
sizes = [315, 90]

# Whether to 'explode' a slice of the pie, and by how much.
explode = [0, 0.1]

# Colours for each slice
colors = ['yellow', 'black']

# Make the plot
fig, ax = plt.subplots()

# Plot the pie chart
handles,_,_ = ax.pie(
    x=sizes, 
    explode=explode, 
    labels=None, 
    autopct='',
    shadow=False, 
    startangle=45, 
    colors=colors
)

# Equal aspect ratio ensures that pie is drawn as a circle.
ax.axis('equal')  

# Add legend
ax.legend(handles, labels, loc='best')

# Set title
ax.set_title('Pacman');

fig.savefig('pacman.jpg')
```
