```python
import glob

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#%matplotlib inline
sns.set_theme(context='notebook', style='darkgrid')

```

```python
files = glob.glob('/Users/jtm545/Projects/Enviroplus/data/*.csv')
files
```

```python
files = glob.glob('/Users/jtm545/Projects/Enviroplus/data/*.csv')
df = []
for file in files:
    data = pd.read_csv(file)
    df.append(data)
    
df = (
    pd.concat(df)
    .loc[lambda df_: df_.timestamp!='timestamp']
    .assign(timestamp=lambda df_: pd.DatetimeIndex(df_.timestamp))
    .set_index('timestamp')
    .sort_index()
    .astype('float')
)
df
```

```python
gas_cols = ['oxidised', 'reduced', 'nh3']
df.groupby(df.index.hour)[gas_cols].mean().plot(subplots=True)
```

```python
fig, ax = plt.subplots(figsize=(12, 4))
ax2 = ax.twinx()
df.groupby(df.index.hour)['humidity'].mean().plot(ax=ax, c='tab:orange', ylabel='Humidity', grid=False)
df.groupby(df.index.hour)['pressure'].mean().plot(ax=ax2, c='tab:blue', ylabel='Humidity', grid=False)
```

```python
gas_cols = ['pm1', 'pm2_5', 'pm10']
df.groupby(df.index.hour)[gas_cols].mean().plot(subplots=True)
```
