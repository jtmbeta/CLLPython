The Finale
----------

How about a lolcat fortune for every cowfile? We can list all available cowfiles using the `-l` option, and thanks to excellent shell integration of Jupyter notebooks we can capture the output into a Python variable.  

```python
cowfiles = !cowsay -l
cowfiles
```

We just want a list of all possible options so let's do a bit of processing

```python
cowfiles = [c.split(' ') for c in cowfiles[1:]]  # First line is not an option
cowfiles = [c for sublist in cowfiles for c in sublist]  # Flatten the list
print(cowfiles)  # This is our list of cowfiles
```

Great, now let's iterate over the cowfiles and execute a shell command for each of them. Again, because Jupyter is closely integrated with the shell, we can affix the loop variable with a dollar sign to pass in the cowfile option when executing the system command. 

```python
for c in cowfiles:
    print(f'~~~ {c} ~~~')
    !fortune | cowsay -f $c | lolcat
```
